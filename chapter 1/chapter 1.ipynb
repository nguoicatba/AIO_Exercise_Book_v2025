1. Viết hàm tính độ đo F1-Score thường được sử dụng để đánh
giá mô hình phân loại.


```python
def evaluate_f1_components(tp, fp, fn):
    # Validate type
    if not all(isinstance(x, int) and type(x) is int for x in [tp, fp, fn]):
        print("Error: tp, fp, fn must be integers (not bool, not float).")
        return

    # Validate non-negative
    if tp < 0 or fp < 0 or fn < 0:
        print("Error: tp, fp, fn must be non-negative integers.")
        return

    # Avoid division by zero
    if tp + fp == 0:
        print("Error: Precision undefined because tp + fp = 0.")
        return
    if tp + fn == 0:
        print("Error: Recall undefined because tp + fn = 0.")
        return

    # Compute
    Precision = tp / (tp + fp)
    Recall = tp / (tp + fn)
    F1 = 2 * Precision * Recall / (Precision + Recall) if (Precision + Recall) != 0 else 0.0

    print("Precision:", Precision)
    print("Recall:", Recall)
    print("F1-score:", F1)

evaluate_f1_components(tp=2, fp=3, fn=4)
evaluate_f1_components(tp="a", fp=3, fn=4)
```
