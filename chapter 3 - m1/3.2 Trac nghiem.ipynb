```python
class Ward:
    def __init__(self, name: str):
        self.__name = name
        self.__listPeople = list()

    def addPerson(self, person: Person):
        self.__listPeople.append(person)

    def describe(self):
        print(f"Ward Name: {self.__name}")
        for p in self.__listPeople:
            p.describe()

    def countDoctor(self):
        ### Your Code Here
        ans =0
        for p in self.__listPeople:
            if (isinstance(p, Doctor)):
                ans += 1
        return ans
        ### End Code Here


student1 = Student(name="studentA", yob=2010, grade="7")
teacher1 = Teacher(name="teacherA", yob=1969, subject="Math")
teacher2 = Teacher(name="teacherB", yob=1995, subject="History")
doctor1 = Doctor(
    name="doctorA",
    yob=1945,
    specialist="Endocrinologists"
)
doctor2 = Doctor(
    name="doctorB",
    yob=1975,
    specialist="Cardiologists"
)

ward1 = Ward(name="Ward1")
ward1.addPerson(student1)
ward1.addPerson(teacher1)
ward1.addPerson(teacher2)
ward1.addPerson(doctor1)
ward1.addPerson(doctor2)

print(ward1.countDoctor())
```

```python
class MyStack:
    def __init__(self, capacity):
        self.__capacity = capacity
        self.__stack = []

    def isFull(self):
        return len(self.__stack) == self.__capacity

    def push(self, value):
        ### Your Code Here
        self.__stack.append(value)
        self.__capacity +=1
        ### End Code Here


stack1 = MyStack(capacity=5)
stack1.push(1)
stack1.push(2)
print(stack1.isFull())
```
